#### 一、什么是哈希表？

哈希表是根据关键码的值而直接进行访问的数据结构。比如数组就是哈希表
![](../attachments/数组就是一张哈希表.png)

#### 二、哈希表用来干嘛？

用来快速判断一个元素是否出现集合里。你像查一个学生在不在这个学校里，枚举是0(n)用哈希表的话就是O(1)

#### 三、哈希函数是干嘛的？

哈希函数的本质是将任意长度的输入数据转换为固定长度的唯一输出（哈希值），以便快速索引和验证数据完整性。![](../attachments/哈希函数的本质.png)


#### 四、为什么会有哈希碰撞？
![](../attachments/哈希冲突.png)

- [哈希函数的输出空间有限]哈希碰撞的产生源于**“鸽巢原理”（Pigeonhole Principle）**，该原理指出，当你有n个“鸽子”和少于n个的“鸽巢”时，总会有至少一个鸽巢包含多于一个鸽子。
- [不同的输入映射到相同的哈希值]由于哈希函数是将一个很大的输入空间映射到一个相对较小的输出空间，这样的映射必然会存在多个输入映射到相同的输出。这种现象称为哈希碰撞。哈希函数通过复杂的计算来尽量使不同的输入数据产生不同的哈希值，但它不能彻底避免碰撞，只能降低碰撞发生的概率。
- 例子：SHA-256 SHA-256 的可能值是 $2^{256}$，因为 SHA-256 的输出是一个 256 位的二进制数，也就是说它的哈希值由 256 个比特位（bit）组成，每个位都可能是 **0** 或 **1**。

具体来说：

- 每个比特位有 2 个可能的状态（0 或 1）。
- 256 位的组合意味着每个比特位可以独立地取这两个状态。
- 因此，可能的组合总数为 $2 \times 2 \times 2 \ldots$ 共 256 次，这就等于 $2^{256}$。

然而，输入可以是任意长度的字符串，这意味着可能存在无限种输入....有限干不过无限的

#### 五、怎么解决哈希碰撞？

##### 5.1、拉链法
> 在拉链法中，每个哈希表的槽位（bucket）实际上是一个链表。当两个或多个元素被映射到相同的位置时，后续的元素就会添加到链表中，形成一个“链式结构”。
![](../attachments/哈希冲突-拉链法.png)


##### 5.2、线性探测法
> 每个槽位只存储一个元素，发生冲突时继续查找表中下一个位置，直到找到空位。所以一定要保证tableSize大于dataSize
![](../attachments/哈希冲突-线性探测法.png)



#### 六、总结

当我们遇到了要快速判断一个元素是否出现集合里的时候，就要考虑哈希法。

但是哈希法也是牺牲了空间换取了时间，因为我们要使用额外的数组，set或者是map来存放数据，才能实现快速的查找。

如果在做面试题目的时候遇到当我们需要查询一个元素是否出现过，或者一个元素是否在集合里的时候，就要第一时间想到哈希法。
